```python
merge1 = pd.merge(df1,df2,on = "Employee_ID")
merge1
```

```python
HR = merge1[merge1["Department"]=="HR"]

Finance = merge1[merge1["Department"]=="Finance"]

Sales = merge1[merge1["Department"]=="Sales"]

It = merge1[merge1["Department"]=="IT"]
```

```python
It = [It["Status"].value_counts()[0] , It["Status"].value_counts()[1] , It["Status"].value_counts()[2]]
It_labels = ["Present" , "Absent" , "Leave"]

HR = [HR["Status"].value_counts()[0] , HR["Status"].value_counts()[1] , HR["Status"].value_counts()[2]]
HR_labels = ["Present" , "Absent" , "Leave"]

Finance = [Finance["Status"].value_counts()[0] , Finance["Status"].value_counts()[1] , Finance["Status"].value_counts()[2]]
Finance_labels = ["Present" , "Absent" , "Leave"]

Sales = [Sales["Status"].value_counts()[0] , Sales["Status"].value_counts()[1] , Sales["Status"].value_counts()[2]]
Sales_labels = ["Present" , "Absent" , "Leave"]
```

```python
import matplotlib.pyplot as plt

fig , ax = plt .subplots(2,2)

ax[0,0].pie(It , labels = It_labels , autopct = "%.2f")
ax[0,0].set_title("IT Attendence Rates")

ax[0,1].pie(HR , labels = HR_labels , autopct = "%.2f")
ax[0,1].set_title("HR Attendence Rates")

ax[1,0].pie( Finance, labels = Finance_labels , autopct = "%.2f")
ax[1,0].set_title("Finance Attendence Rates")

ax[1,1].pie(Sales , labels = Sales_labels , autopct = "%.2f")
ax[1,1].set_title("Sales Attendence Rates")
```

```python
import matplotlib.pyplot as plt

departments = ["IT", "HR", "Finance", "Sales"]

fig, axes = plt.subplots(2, 4, figsize=(14, 7))
axes = axes.flatten()

for i, dept in enumerate(departments):
    df = merge1[merge1["Department"] == dept]
    status_counts = df["Status"].value_counts()
    hours_worked = df["Hours_Worked"].sum()

    # Attendance Pie
    data_att = [status_counts.get(s, 0) for s in ["Present", "Absent", "Leave"]]
    labels_att = ["Present", "Absent", "Leave"]
    axes[i].pie(data_att, labels=labels_att, autopct='%1.1f%%', startangle=90)
    axes[i].set_title(f"{dept} Attendance")

    # Hours Pie (Worked vs Remaining)
    total_possible = df.shape[0] * 8  
    data_hours = [hours_worked, total_possible - hours_worked]
    labels_hours = ["Worked Hours", "Remaining"]
    axes[i+4].pie(data_hours, labels=labels_hours, autopct='%1.1f%%', startangle=90)
    axes[i+4].set_title(f"{dept} Hours Worked")

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sirf Leave rows filter karo
leave_df = merge1[merge1["Status"] == "Leave"]

# Department-wise Leave counts
leave_counts = leave_df["Department"].value_counts()

# Pie chart
plt.figure(figsize=(6,6))
plt.pie(leave_counts, labels=leave_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Department-wise Leave Breakdown")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


total_days = merge1.groupby("Employee_ID")["Status"].count()
present_days = merge1[merge1["Status"] == "Present"].groupby("Employee_ID")["Status"].count()
attendance_rate = (present_days / total_days).fillna(0)

salary = merge1.groupby("Employee_ID")["Salary"].mean()

df_corr = pd.DataFrame({"Attendance_Rate": attendance_rate,"Salary": salary})

# Correlation
print(df_corr.corr())

# Scatter plot
plt.figure(figsize=(6,6))
plt.scatter(df_corr["Attendance_Rate"], df_corr["Salary"], alpha=0.7)
plt.xlabel("Attendance Rate")
plt.ylabel("Salary (Performance)")
plt.title("Performance vs Attendance Correlation")
plt.grid(True)
plt.show()
```
