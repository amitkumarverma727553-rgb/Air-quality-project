# Q1

```python
country = list(df["Headquarters"])
print(country)
```

```python
n_america = ["United States" , "Canada" , "Mexico"]
s_america = ["Brazil" , "Argentina"]
asia = ["India" , "China" , "Japan" , "South Korea" , "Taiwan" , "Hong Kong" , "Indonesia" , "Singapore" , "Thailand" , "Malaysia"]
middle_east = ["Saudi Arabia" , "Qatar" , "Iran" , "Isreal" , "United Arab Emirates"]
australia = ["Australia"]
africa = ["South Africa"]
europe = []

continent = []

for i in country:
    if i in n_america:
        continent.append("North America")
    elif i in s_america:
        continent.append("South America")
    elif i in asia:
        continent.append("Asia")
    elif i in middle_east:
        continent.append("Middle East")
    elif i in australia:
        continent.append("Australia")
    elif i in africa:
        continent.append("Africa")
    else:
        continent.append("Europe")
```

```python
print(continent)
```

```python
df["Continent"] = continent
```

```python
df
```

```python
df.info()
```

```python
profit = list(df["Profit ($B)"])
pro = []
for i in profit:
    if i.count(".") > 1:
        pro.append(0)
    else:
        pro.append(float(i))
```

```python
print(pro)
```

```python
df["Profit ($B)"]  = pro
```

```python
df
```

```python
df.info()
```

```python
df["Profit Margin"] = df["Profit ($B)"] / df["Sales ($B)"]
df
```
