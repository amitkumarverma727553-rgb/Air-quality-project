```python
import pandas as pd
df1 = pd.read_excel(r"C:\Users\amitk\OneDrive\Documents\project2_ecommerce_data\customers.xlsx")
df1
```

```python
df2 = pd.read_excel(r"C:\Users\amitk\OneDrive\Documents\project2_ecommerce_data\orders.xlsx")
df2
```

```python
df3 = pd.read_csv(r"C:\Users\amitk\OneDrive\Documents\project2_ecommerce_data\products.csv")
df3
```

```python
merge1 = pd.merge(df2,df3,on = "Product_ID")
merge1
```

```python
import pandas as pd
import matplotlib.pyplot as plt

top_selling = merge1[merge1["Status"]=="Delivered"].groupby("Product_Name")["Quantity"].sum().sort_values(ascending=False).head(5)

top_returned = merge1[merge1["Status"]=="Returned"].groupby("Product_Name")["Quantity"].sum().sort_values(ascending=False).head(5)

fig, axes = plt.subplots(1, 2, figsize=(12,5))
top_selling.plot(kind="bar", ax=axes[0], color="skyblue")
axes[0].set_title("Top-Selling Products")
axes[0].set_ylabel("Quantity Sold")

top_returned.plot(kind="bar", ax=axes[1], color="salmon")
axes[1].set_title("Top-Returned Products")
axes[1].set_ylabel("Quantity Returned")

plt.tight_layout()
plt.show()
```

```python
merge1["Date"] = pd.to_datetime(merge1["Date"])

monthly_orders = (merge1[merge1["Status"]=="Delivered"].groupby(merge1["Date"].dt.month)["Customer_ID"].count())

plt.figure(figsize=(8,5))
monthly_orders.plot(kind="bar", color="skyblue", edgecolor="black")
plt.title("Seasonal Order Spikes (Monthly)")
plt.xlabel("Month")
plt.ylabel("Number of Orders")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

merge1["Revenue"] = merge1["Quantity"] * merge1["Unit_Price"]

category_revenue = merge1.groupby("Category")["Revenue"].sum()

plt.figure(figsize=(6,6))
category_revenue.plot(kind="pie",autopct="%.1f%%",startangle=90,shadow=True,cmap="Set3")
plt.title("Category Contribution to Revenue")
plt.ylabel("")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

merge1["Date"] = pd.to_datetime(merge1["Date"])

merge1["Revenue"] = merge1["Quantity"] * merge1["Unit_Price"]

monthly_inflow = merge1[merge1["Status"]=="Delivered"].groupby(merge1["Date"].dt.to_period("M"))["Revenue"].sum()

monthly_outflow = merge1[merge1["Status"]=="Returned"].groupby(merge1["Date"].dt.to_period("M"))["Revenue"].sum()

monthly_cashflow = monthly_inflow - monthly_outflow

plt.figure(figsize=(10,6))
monthly_inflow.plot(label="Inflow (Sales)", marker="o")
monthly_outflow.plot(label="Outflow (Returns)", marker="o", color="red")
monthly_cashflow.plot(label="Net Cash Flow", marker="o", color="green")

plt.title("Monthly Cash Flow Trends")
plt.xlabel("Month")
plt.ylabel("Amount")
plt.legend()
plt.grid(True)
plt.show()
```

```python
merge1["Revenue"] = merge1["Quantity"] * merge1["Unit_Price"]

merge1["Date"] = pd.to_datetime(merge1["Date"])

def get_season(month):
    if month in [12,1,2]:
        return "Winter"
    elif month in [3,4,5]:
        return "Spring"
    elif month in [6,7,8]:
        return "Summer"
    else:
        return "Autumn"

merge1["Season"] = merge1["Date"].dt.month.apply(get_season)

seasonal_revenue = merge1.groupby("Season")["Revenue"].sum().reindex(["Winter","Spring","Summer","Autumn"])

plt.figure(figsize=(8,5))
seasonal_revenue.plot(kind="bar", color="teal")
plt.title("Seasonal Revenue Growth")
plt.ylabel("Total Revenue")
plt.xlabel("Season")
plt.show()
```
